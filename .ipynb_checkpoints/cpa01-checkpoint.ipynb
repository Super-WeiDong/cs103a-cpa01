```python
import pandas as pd
from matplotlib import pyplot as plt
#pd.set_option('display.float_format',lambda x : '%.2f' % x)
```

# Reading in the data!
We use the read_csv function to create our data frames

```python
df = pd.read_csv('life-expectancy-at-birth-total-years.csv')
```

# rough overview of the data

```python
df.describe()
```

# print the array of columns and the index array

```python
df.index
```

```python
df.columns
```

# create some simple plot of part of the data

```python
ch = df[(df['Entity'] == 'China')]
us = df[(df['Entity'] == 'United States')]
C = ch['Life expectancy at birth, total (years)']
U = us['Life expectancy at birth, total (years)']
X = ch['Year']
plt.plot(X,C,label='CHA')
plt.plot(X,U,label='US')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title("Life Expectancy comparasion between China and US")
plt.legend()
plt.grid()
```

# create a pivot table and plot some data from that pivot table to answer Q1
Q1: How life expectancy has changed in the U.S., China,Russia,Ukerain from 1960 -2019?

```python
df2 = pd.pivot_table(df, values="Life expectancy at birth, total (years)", index=["Year"], columns=["Entity"])
df2
```

```python
df2[['China','United States','Russia','Ukraine']].plot(figsize=(20,10))
plt.grid()
```

At beginning, life expectancy of china lags far behind the other three countries.

Later,life expectancy of china has continued to grow rapidly, and surpassed Russia and Ukraine in 1990s

Now, China's life expectancy is approaching that of the US.

# use the groupby feature to answer Q2
Q2: How average life expectancy has changed in the our world from 1960 -2019?

```python
df4 = df.groupby('Year')
df4['Life expectancy at birth, total (years)'].agg(['mean','max','min','median']).plot(figsize=(20,10))
plt.grid()
```
